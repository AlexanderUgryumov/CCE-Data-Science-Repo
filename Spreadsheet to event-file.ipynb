### Excel spreadsheet to event file to load to Petrel

```python
import pandas as pd

df = pd.read_excel('Perf to event file2.xlsx')



Well_list = ["130_h", "134_h", "136_h", "138_h", "152_h", "157_h" ,"162_h" ,"200_h" ,"201_h" , "202_h",
             "204_h", "205_h" ,"206_h" ,"207_h", "208_h" ,"209_h" ,"211_h" ,"212_h", "213_h", "214_h",
             "216_h", "218_h" , "219_h", "221_h", "223_h", "227_h", "240_h", "4252_h", "4253_h"]

for well in Well_list:
    file_name = well + '.ev'
    with open(file_name, 'w') as file:
        file.write('UNITS METRIC \n \nWELLNAME ')
        file.write("\"" + str(well) + "\"")
        file.write('\n')
    df_well = df[(df['Well'] == well)]
    df_well = df_well.drop(['Well'], axis=1)
    df_well.to_csv(file_name, sep='\t', mode='a+', header=False, index=False)
df.head()
```
