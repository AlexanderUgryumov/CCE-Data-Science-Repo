```python
def RT_parce(file_in, file_out, N):
    """
    This function is intended to parse core resistivity from the Neftecom standard file into csv file loadable into Techlog
    Arguments:
        file_in <string> fine name which needs to be parsed. Needs to be in 'xlsx' format
        file_out <string> name of the output file
        number of rows in the source spreadsheet
    """
    from openpyxl import load_workbook # import lib to process excel files
    wb2 = load_workbook(file_in)
    sheet = wb2.active

    
    # Create an output file and write into it curves headers and units
    header = 'SampleID; MD; CPOR; CKKL; CFRF; CRI; CSW;\n TEXT; m; %; mD; v/v; v/v; %\n'

    with open(file_out, 'w') as file1:
        for line in header:
            file1.write(line)
        
    SampleID = []
    # DIR = []
    MD = []
    Poro = []
    Perm = []
    CFRF = []
    CRI = []
    CSW = []

    
    for row in sheet.iter_rows(min_row = 7, min_col = 1, max_row = N, max_col = 24, values_only = True):
            SampleID.append(row[1])
            # DIR.append(row[2])
            top_depth = row[3]
            position = row[6]
            MD.append(top_depth + position)
            Poro.append(row[8])
            Perm.append(row[9])
            CFRF.append(row[11])
            CRI.extend([row[14], row[17], row[20]])
            CSW.extend([row[13], row[16], row[19]])
    

    file1 = open(file_out, 'a+')

    
    # Write data into output file suitable to load into Techlog as array data
    for i in range(len(SampleID)):
        for j in range(i*3,(i+1)*3):
            row = map(str,[SampleID[i], MD[i], Poro[i], Perm[i], CFRF[i], CRI[j], CSW[j]])
            fin = ';'.join(row)
            fin1 = fin.replace('None', '-9999')
            file1.writelines(fin1)
            file1.write('\n')
    file1.close()

file_in = 'УЭС_611.xlsx'
file_out = '611R_Resistivity.csv'

RT_parce(file_in, file_out, 94)
```
